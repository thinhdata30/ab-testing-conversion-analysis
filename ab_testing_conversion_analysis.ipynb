```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/quang/Downloads/marketing_AB.csv")
df.head()
```

```python
df['converted'] = df['converted'].astype(int)
```

```python
df.groupby('test group')['converted'].mean()
```

```python
from statsmodels.stats.proportion import proportions_ztest

ad = df[df['test group'] == 'ad']['converted']
psa = df[df['test group'] == 'psa']['converted']

proportions_ztest([ad.sum(), psa.sum()], [len(ad), len(psa)])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='test group', y='converted', data=df)
plt.show()
```
